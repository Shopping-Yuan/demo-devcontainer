```python
path_list = glob.glob("/workspaces/demo-devcontainer/res_gossiping/*.txt")
```

```python
article_list = []
columns = []
for index, path in enumerate(path_list):
    with open (path,'r') as f:
        article_string = f.read()
    article_info = article_string.split("---split---")[-1].split("\n")[1:-1]
    if index == 0:
        columns = [x.split(":")[0] for x in article_info]
        article_dataframe = pd.DataFrame(
            columns = columns
        )
    # Text to list-element in DataFrame
    article_info = [x.split(":")[-1] for x in article_info]
    article_dataframe.loc[index] = article_info
article_dataframe
```
